```python
'''7.'''

clientes = [] # cria variavel que será preenchida com os inputs


while True:
    numero = int(input("Insira o ID do cliente: ")) # input que espera um n° inteiro 
    
    clientes.append(numero) # uso do append para adicionar a lista 'clientes'

    continuar = input("Deseja inserir outro número? (S/N): ").upper() # pede se deve continuar ou parar os inputs

    if continuar != 'S': # se não for 'S', parar de pedir input
        break
        
# a) todos os números digitados na ordem em que foram inseridos
print("\nTodos os números digitados na ordem em que foram inseridos:") 
print(clientes) #como usei append, ele coloca o n° em ordem que foi adicionado

# b) todos os números digitados em ordem crescente
print("\nTodos os números digitados em ordem crescente:")
clientes_ordenados = sorted(clientes) # uso do sorted para ordenar em forma crescente 
print(clientes_ordenados)

# c) a média destes valores
media = sum(clientes) / len(clientes) # formula da media, usando 'sum' + 'len' para facilitar/simplificar
print(f"\nA média dos valores digitados é: {media:.2f}") # uso do f-string novamente para escolher qtd de casa decimais

# d) apenas os números pares
print("\nApenas os números pares:")
numeros_pares = [num for num in clientes if num % 2 == 0]
print(numeros_pares) # for percorre numeros na lista e se é divisivel por 2 e o resto da divisão for '0' cria uma nova lista de n°s pares

# e) apenas os números ímpares
print("\nApenas os números ímpares:")
numeros_impares = [num for num in clientes if num % 2 != 0] 
print(numeros_impares) # mesma lógica acima, porém fazendo o uso do diferente "!=0" e também cria nova lista para os ímpares

# f) apenas os números repetidos
print("\nApenas os números repetidos:")
numeros_repetidos = [] # cria variavel vazia de lista

for num in clientes:
    # e se ainda não foi adicionado à lista de números repetidos
    if clientes.count(num) > 1 and num not in numeros_repetidos: #verifica se o n° é repetido e se já n foi adicionao
        numeros_repetidos.append(num) # adicionando com append a variavel exclusiva criada

print(numeros_repetidos)
```

```python
'''8. '''

# inputs conforme pedido no enunciado (usando upper para letra maiuscula):
temperatura = float(input("Digite a temperatura: "))

escala_original = input("Digite a temperatura atual (C, K ou F): ").upper()

escala_convertida = input("Digite a temperatura para converter: (C, K ou F): ").upper()

# usando 'if' (se) e 'elif' (sempre que um não der, tentar o outro) 

if escala_original == 'C': # de celsius para kelvin
    if escala_convertida == 'K':
        temperatura = temperatura + 273.15
    elif escala_convertida == 'F': # se nao, para fahrenheit 
        temperatura = (temperatura * 9/5) + 32
elif escala_original == 'K':
    if escala_convertida == 'C': # de kelvin para celsiues
        temperatura = temperatura - 273.15
    elif escala_convertida == 'F': # se nao, para fahrenheit 
        temperatura = (temperatura - 273.15) * 9/5 + 32
elif escala_original == 'F': # de fahrenheit para celsius....
    if escala_convertida == 'C': 
        temperatura = (temperatura - 32) * 5/9
    elif escala_convertida == 'K':
        temperatura = (temperatura - 32) * 5/9 + 273.15

print(f"A temperatura escolhida é: {temperatura:.2f} {escala_convertida}") 
# uso do f-string novamente para escolher qtd de casa decimais (.2f)

```

```python
'''9. '''

produtos = ["Crediário", "Consórcio", "Consórcio", "Financiamento", "Consórcio", "Crediário"]
produtos_ordenados = sorted(produtos)  # o 'sorted' ajusta a lista para ordem alfabetica

while True:
    novo_produto = input("Digite o produto (ou 'sair' para finalizar): ")

    if novo_produto.lower() == 'sair':
        break  # condição que quebra o loop se for digitado 'sair'

    produtos_ordenados.append(novo_produto)  # usando 'append' para adicionar novo produto

produto_atual = produtos_ordenados[0]
quantidade_vendas = 1

for produto in produtos_ordenados[1:]: # for para percorrer sobre os produtos
    if produto == produto_atual:
        quantidade_vendas += 1  # adc +1 a quantidade de vendas se o nome do produto for igual
    else:
        print(f"{produto_atual} - {quantidade_vendas}")  # imprime nome do produto + qtd de vendas
        produto_atual = produto
        quantidade_vendas = 1  # reinicia a contagem

print(f"{produto_atual} - {quantidade_vendas}") 

```

```python
'''10. '''

# lista conforme enunciado
valores = [156.89, 155.16, 148.41, 145.18, 150.23, 148.10, 155.68,
    146.07, 149.53, 151.67, 158.16, 150.09, 145.64, 155.12,
    152.37, 145.01, 158.19, 159.66, 156.20, 158.04, 146.20,
    154.60, 157.98, 153.68, 149.44, 142.01, 148.68, 152.22, 158.26, 159.33]

media_movel = [] # variavel vazia para armezanar os resultados depois

for i in range(len(valores) - 6): # for para percorrer os valores e calcular a media movel de 7d para os periodos
    # uso do len para pegar o tamanho da lista - 6, pra ter certeza que restem 7 elementos a partir do indice 'i'
    media = sum(valores[i:i+7]) / 7  # calculo da media dos valores de fechamento -> obs: usando [i:i+7] para fatiar e pegar oq corresponde ao periodo de 7d da lista 'valores'
    media_movel.append(media)  # usando 'append' para adicioanr a media a lista 

print("\nValores de fechamento:\n", valores)
print("\nMedia movel de 7 dias:\n", media_movel)


```
